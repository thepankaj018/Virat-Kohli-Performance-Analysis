let’s have a look at the trend of runs scored by Virat Kohli in his career from 18 August 2008 to 22 January 2017:

```python
plt.figure(figsize=(16,5))
sns.lineplot(x=df.index,y='Runs',data=df)
plt.title('Runs Scored by Virat Kohli Between 18-Aug-08 - 22-Jan-17')
plt.show()
```

```python
df.head()
```

```python
df['Pos'].value_counts()
```

let’s see all the batting positions played by Virat Kohli:

```python
plt.figure(figsize=(16,5))
ax=sns.countplot(x='Pos',data=df)
ax.bar_label(ax.containers[0])
plt.title('Number of Matches At Different Batting Positions')
plt.show()
```

```python
df.head()
```

```python
pos_runs=df.groupby('Pos')['Runs'].sum().reset_index()
```

```python
pos_runs
```

```python
plt.style.use('fivethirtyeight')
```

let’s have a look at the total runs scored by Virat Kohli in different positions:

```python
plt.figure(figsize=(16,5))
ax=sns.barplot(x='Pos',y='Runs',data=pos_runs)
ax.bar_label(ax.containers[0])
plt.title('Runs Scored at different Positions')
plt.show()
```

Now let’s have a look at the number of centuries scored by Virat Kohli while batting in the first innings and second innings

```python
centuries=df.query('Runs>=100')
```

```python
centuries
```

```python
plt.figure(figsize=(18,5))
ax=sns.countplot(x='Inns',data=centuries,hue='Opposition')
ax.bar_label(ax.containers[0])
plt.title('Centuries By Virat Kohli in First Innings Vs. Second Innings')
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.show()
```

Now let’s have a look at the kind of dismissals Virat Kohli faced most of the time:

```python
df.head()
```

```python
df.Dismissal.value_counts()
```

```python
plt.figure(figsize=(18,5))
ax=sns.countplot(x='Dismissal',data=df,order=df.Dismissal.value_counts().index)
ax.bar_label(ax.containers[0])
plt.title('Dismissals of Virat Kohli')
plt.show()
```

Now let’s have a look at against which team Virat Kohli scored most of his runs

```python
df.head()
```

```python
runs_scored=df.groupby('Opposition')['Runs'].sum().reset_index()
runs_scored
```

```python
plt.figure(figsize=(18,5))
ax=sns.barplot(x='Opposition',y='Runs',data=runs_scored)
ax.bar_label(ax.containers[0])
plt.title('Most Runs Against Teams')
plt.show()
```

Now let’s have a look at against which team Virat Kohli scored most of his centuries:

```python
centuries.head()
```

```python
centuries['Opposition'].value_counts()
```

```python
plt.figure(figsize=(18,5))
ax=sns.countplot(x='Opposition',hue='Inns',data=centuries)
ax.bar_label(ax.containers[0])
plt.title('Most Centuries Against Teams')
plt.show()
```

analyze Virat Kohli’s strike rate

```python
df.head()
```

```python
strike_rate=df.query("SR>=120")
strike_rate
```

```python
inns_strike_rate=strike_rate.groupby('Inns')['SR'].mean().reset_index()
inns_strike_rate
```

```python
plt.figure(figsize=(18,5))
ax=sns.barplot(x='Inns',y='SR',data=inns_strike_rate)
ax.bar_label(ax.containers[0])
plt.title('Most Centuries Against Teams')
plt.show()
```

total number of 4's and 6's against each team

```python
boundary=df.groupby('Opposition')[['4s','6s']].sum().reset_index()
boundary
```

```python
plt.figure(figsize=(18,5))
ax=sns.barplot(x='Opposition',y='4s',data=boundary)
ax.bar_label(ax.containers[0])
plt.title('Most boundary Against Teams')
plt.show()
```

```python
plt.figure(figsize=(18,5))
ax=sns.barplot(x='Opposition',y='6s',data=boundary)
ax.bar_label(ax.containers[0])
plt.title('Most boundary Against Teams')
plt.show()
```
